```python
import pandas as pd
import numpy as np
import scipy as sp
import statsmodels.api as sm
import matplotlib.pyplot as plt
from datetime import datetime
from dateutil.parser import parse
```

```python
with open(r"C:\Users\abhij\Desktop\UVa Coursework\Capstone\Conn logs\conn_malacious.log") as f:
    logs = pd.Series(f.readlines())
```

```python
cols = logs[6].split("\t")[1:]
cols
```

```python
df = logs[8:-1].str.split('\t', expand=True)
df.columns = cols
df.reset_index()
```

```python
df['ts'].astype(float)
```

```python
trace_df = df[['id.orig_h',"id.resp_h", "id.resp_p", "proto"]].drop_duplicates().reset_index(drop = True).reset_index()
trace_df.rename(columns = {"index":"trace_id"}, inplace = True)
```

```python
trace_df
```

```python
df_merge = df.merge(trace_df, how = "left", on =['id.orig_h',"id.resp_h", "id.resp_p", "proto"])
```

```python
df_merge['ts'] = df_merge['ts'].astype(object).astype(float)
df_merge = df_merge.sort_values(by=["trace_id","ts"]).reset_index(drop = True)
```

```python
df_merge.loc[1:,"time_interval"] = df_merge.loc[1:,"ts"].reset_index(drop = True) - df_merge.loc[:,"ts"].head(df_merge.shape[0])
```

```python
df_merge.head(30)
```

```python
df_merge["last_trace_index"] = df_merge.index
```

```python
last_index_list = df_merge.groupby("trace_id").last().last_trace_index
```

```python
df_merge.loc[last_index_list,"time_interval"] = -1
```

```python
average_time_interval = pd.pivot_table(df_merge.loc[df_merge['time_interval']>=0,:], values = 'time_interval', index = 'trace_id', aggfunc = np.mean)
```
